1.Importing the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

2.Load the dataset

```python
dataset = pd.read_csv("/Users/user/Documents/Fraud_detection/Churn_Modelling.csv")
dataset
```

3.Feature selection

```python
X = dataset.iloc[:, 3:13].values
Y = dataset.iloc[:,-1].values
```

4.Encoding the Dataset

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
X[:,1] = labelencoder_X_1.fit_transform(X[:,1])
labelencoder_X_2 = LabelEncoder()
X[:,2] = labelencoder_X_2.fit_transform(X[:,2])
onehotencoder = OneHotEncoder(categorical_features= [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:,1:]
```

5.Splitting the dataset into train and test set

```python

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

6.Normalizing the dataset(Feature Scaling)

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

7.Build the ANN using keras



```python
import keras
from keras.models import Sequential #for initializing ANN models
from keras.layers import Dense

#Initializing ANN
classifier = Sequential()

#Forward propogation
#Adding layers #First hidden layer
classifier.add(Dense(output_dim = 6, init = 'random_uniform', activation= 'relu', input_dim = 11))

#Adding second hidden layer
classifier.add(Dense(output_dim = 6, init = 'random_uniform', activation= 'relu'))

#Adding output layer
classifier.add(Dense(output_dim = 1, init = 'random_uniform' ,activation= 'sigmoid'))

```

8.Calculate loss and update parameters

```python
classifier.compile(optimizer= 'adam', loss= 'binary_crossentropy', metrics = ['accuracy'])
```

```python
y_pred = classifier.predict(X_train)
y_pred = (y_pred > 0.5)
```

9.Predict the accuracy 

```python
classifier.fit(X_train, Y_train, batch_size= 20, epochs = 100)
```

```python
classifier.fit(X_test, Y_test, epochs = 100)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
```
